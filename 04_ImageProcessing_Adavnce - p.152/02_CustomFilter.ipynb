#### 02_Custom Filter 적용하기

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
```

```python
image_path = '../99_Studyfile/02_Sample/'   # 폴더 지정

image_list = glob.glob(os.path.join(image_path, '*.jpeg'))  # 지정 이미지 형식 모두 불러오기
```

```python
# filter2D 효과

for path in image_list:
    print(path)
    image = cv2.imread(path, 1)
    image = cv2. cvtColor(image, cv2.COLOR_BGR2RGB)
    
    
    # Custom Filter
    filter = np.array([[-1,3,3],
                       [-3,7,-5],
                       [5,-1,-6]])
    
    # Custom Filter 적용
    custom_filter_img = cv2.filter2D(image, -2, filter)
    
    plt.imshow(custom_filter_img)
    plt.show()
```

#### 다양한 효과 필터

```python
# Sepia 효과

for path in image_list:
    print(path)
    image = cv2.imread(path, 1)
    
    # 효과 필터 적용
    filter_ = np.array([[0.572, 0.734, 0.872],[0.621, 0.826, 0.443],[0.124, 0.145, 0.166]])
    
    sepia_img = cv2.transform(image, filter_)
    
    plt.imshow(sepia_img)
    plt.show()
```

#### 엠보스 효과

```python
# embossing 효과

for path in image_list:
    print(path)
    image = cv2.imread(path, 1)
    
    # 효과 적용
    filter_temp = np.array([[0,1,0],[0,0,0],[0,-1,0]])
    
    emboss_img = cv2.filter2D(image, -1, filter_temp)
    emboss_img = emboss_img + 100
    
    plt.imshow(emboss_img)
    plt.show()
```
